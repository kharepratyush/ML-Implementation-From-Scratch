```python
import pandas as pd
import seaborn as sns
from numpy.linalg import inv
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import numpy as np
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import random
import statistics 
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
sns.set()
```

## Linear Regression - Linear Algebra

```python
def linear_regression_linear_algebra(orig_dataset, features, target):
    dataset = orig_dataset.copy()
    dataset['coef'] = 1
    #print(dataset.head(2))
    features = (list(features)+['coef'])
    H = dataset[features].values
    y = dataset[target].values
    
    try:
        H.shape[1]
    except:
        H = H.reshape((len(H), 1))
        y = y.reshape((len(y), 1))
        
    w = inv(H.T.dot(H)).dot(H.T).dot(y)
    y_pred = H.dot(w)
    dataset['y_pred'] = y_pred
    
    print('------------------------------------------------')
    print('R2 Score: ', r2_score(dataset[target], dataset['y_pred']))
    print('MSE: ', mean_squared_error(dataset[target], dataset['y_pred']))
    print('RMSE: ', np.sqrt(mean_squared_error(dataset[target], dataset['y_pred'])))
    print('MAE: ', mean_absolute_error(dataset[target], dataset['y_pred']))
    print('------------------------------------------------')
    
    print('Features: ', features)
    print('Coefficient:', w)
    return dataset.drop('coef', axis=1), w
```

### Low Dimension Data

```python
dataset = pd.read_csv('01. Linear Regression Dataset.csv')
dataset.dropna(inplace=True)
```

```python
scaler = StandardScaler()
dataset = pd.DataFrame(scaler.fit_transform(dataset), columns = dataset.columns)
```

```python
sns.scatterplot(x = 'x', y='y', data=dataset)
plt.show()
```

## Gradient Descent

```python
def linear_regression_GD(orig_dataset, features, target, learning_rate = 1e-4, threshold = 1e-8, n_iter = 1e10):
    dataset = orig_dataset.copy()
    dataset['coef'] = 1
    
    features = (list(features)+['coef'])
    H = dataset[features].values
    y = dataset[target].values
    
    try:
        H.shape[1]
    except:
        H = H.reshape((len(H), 1))
        y = y.reshape((len(y), 1))

    w = np.zeros(len(features))

    cost_prev = np.inf
    
    for j in range(int(n_iter)):
        y_pred = H.dot(w)
        
        cost = (1/len(y)) * np.sum(np.square(y_pred - y))
        
        if cost > cost_prev:
            print('Cost increasing - Reducing Learning Rate')
            learning_rate = learning_rate/5
            #print(cost, cost_prev)
            #print(learning_rate, threshold)
        
        #print(learning_rate)
        #print('Cost: ', (1/len(y)) * np.sum(np.square(y_pred - y)))
        
        if abs(cost_prev - cost) < threshold:
            #print(abs(sum(map(lambda x:x*x,w )) - sum(map(lambda x:x*x,w_upd))))
            #print(w_upd)
            break
        
        cost_prev = cost
        w_upd = w - (1/len(y)) * (learning_rate * (H.T.dot(y_pred - y)))
           
        w = w_upd
        
    y_pred = H.dot(w)
    
    dataset['y_pred'] = y_pred
    
    print('------------------------------------------------')
    print('R2 Score: ', r2_score(dataset[target], dataset['y_pred']))
    print('MSE: ', mean_squared_error(dataset[target], dataset['y_pred']))
    print('RMSE: ', np.sqrt(mean_squared_error(dataset[target], dataset['y_pred'])))
    print('MAE: ', mean_absolute_error(dataset[target], dataset['y_pred']))
    print('------------------------------------------------')
    
    print('Features: ', features)
    print('Coefficient:', w)
    return dataset.drop('coef', axis=1), w
```

### Low Dimension Data

```python
%%time
pred_dataset, w = linear_regression_GD(dataset, 'x', 'y')
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x = 'x', y='y', data=pred_dataset, ax=ax)
sns.lineplot(x = 'x', y='y_pred', data=pred_dataset, ax=ax, color='red')
plt.show()
```

## MiniBatch Gradient Descent

```python
def linear_regression_MiniBatchGD(orig_dataset, features, target, 
                                  learning_rate = 1e-4, threshold = 1e-8, n_iter = 1e10, batch_size = 320):
    
    n_iter = n_iter / batch_size
    
    dataset = orig_dataset.copy(deep=True)
    dataset['coef'] = 1
    
    features = (list(features)+['coef'])
    H = dataset[features].values
    y = dataset[target].values
    
    try:
        H.shape[1]
    except:
        H = H.reshape((len(H), 1))
        y = y.reshape((len(y), 1))

    w = np.random.randn(len(features))
    #print(w)
    
    batch_index = [i for i in range(len(y))]
    random.shuffle(batch_index)
    final_shuffled_batches = [batch_index[i * batch_size:(i + 1) * batch_size] for i in range((len(batch_index) + batch_size - 1) // batch_size )]  

    converged = 0
    
    cost_prev = np.inf
    for j in range(int(n_iter)):
        
        cost = 0
        w_prev = w
        
        for batch in final_shuffled_batches:
            H_minibatch = H[batch]
            y_minibatch = y[batch]
            y_minibatch_pred = H_minibatch.dot(w)
            cost += (1/len(y_minibatch)) * np.sum(np.square(y_minibatch_pred - y_minibatch))
            w_upd = w - (1/len(y_minibatch)) * (learning_rate * (H_minibatch.T.dot(y_minibatch_pred - y_minibatch)))
            w = w_upd
            
#         if j % 1000 == 0:
#             print('Cost :', cost)
        
        if cost_prev < cost:
            print('Cost increasing - Reducing Learning Rate')
            learning_rate = learning_rate / 10
            
        if abs(cost_prev - cost) < threshold:
            converged = 1
            break       
            
        cost_prev = cost
        
    y_pred = H.dot(w)
    
    dataset['y_pred'] = y_pred
    
    print('------------------------------------------------')
    print('R2 Score: ', r2_score(dataset[target], dataset['y_pred']))
    print('MSE: ', mean_squared_error(dataset[target], dataset['y_pred']))
    print('RMSE: ', np.sqrt(mean_squared_error(dataset[target], dataset['y_pred'])))
    print('MAE: ', mean_absolute_error(dataset[target], dataset['y_pred']))
    print('------------------------------------------------')
    
    print('Features: ', features)
    print('Coefficient:', w)
    return dataset.drop('coef', axis=1), w_upd
```

### Low Dimension Data

```python
%%time
pred_dataset, w = linear_regression_MiniBatchGD(dataset, 'x', 'y')
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x = 'x', y='y', data=pred_dataset, ax=ax)
sns.lineplot(x = 'x', y='y_pred', data=pred_dataset, ax=ax, color='red')
plt.show()
```

## Ridge Regularization - Using Mini Batch Gradient Descent

```python
def Ridge_regression_MiniBatchGD(orig_dataset, features, target, 
                                  learning_rate = 1e-4, threshold = 1e-8, lambda_ = 1e-2, n_iter = 1e10, batch_size = 320):
    
    n_iter = n_iter / batch_size
    
    dataset = orig_dataset.copy(deep=True)
    dataset['coef'] = 1
    
    features = (list(features)+['coef'])
    H = dataset[features].values
    y = dataset[target].values
    
    try:
        H.shape[1]
    except:
        H = H.reshape((len(H), 1))
        y = y.reshape((len(y), 1))

    w = np.random.randn(len(features))
    #print(w)
    
    batch_index = [i for i in range(len(y))]
    random.shuffle(batch_index)
    final_shuffled_batches = [batch_index[i * batch_size:(i + 1) * batch_size] for i in range((len(batch_index) + batch_size - 1) // batch_size )]  

    converged = 0
    
    cost_prev = np.inf
    for j in range(int(n_iter)):
        
        cost = 0
        w_prev = w
        
        for batch in final_shuffled_batches:
            H_minibatch = H[batch]
            y_minibatch = y[batch]
            y_minibatch_pred = H_minibatch.dot(w)
            cost += (1/len(y_minibatch)) * np.sum(np.square(y_minibatch_pred - y_minibatch))
            
            derivative = (H_minibatch.T.dot(y_minibatch_pred - y_minibatch)) + lambda_*w
            w_upd = w - (1/len(y_minibatch)) * (learning_rate * derivative)
            w = w_upd
            
        #print('Cost :', cost)
        
        if cost_prev < cost:
            print('Cost increasing - Reducing Learning Rate')
            learning_rate = learning_rate / 10
            
        if abs(cost_prev - cost) < threshold:
            converged = 1
            break       
            
        cost_prev = cost
        
    y_pred = H.dot(w)
    
    dataset['y_pred'] = y_pred
    
    print('------------------------------------------------')
    print('R2 Score: ', r2_score(dataset[target], dataset['y_pred']))
    print('MSE: ', mean_squared_error(dataset[target], dataset['y_pred']))
    print('RMSE: ', np.sqrt(mean_squared_error(dataset[target], dataset['y_pred'])))
    print('MAE: ', mean_absolute_error(dataset[target], dataset['y_pred']))
    print('------------------------------------------------')
    
    print('Features: ', features)
    print('Coefficient:', w)
    return dataset.drop('coef', axis=1), w_upd
```

## Lasso Regularization - Coordinate Descent

\begin{aligned}
     \frac{\partial }{\partial \theta_j} RSS^{OLS}(\theta) & = -  \sum_{i=1}^m x_j^{(i)}  \left[y^{(i)} - \sum_{j=0}^n \theta_j x_j^{(i)}\right]
     \\
     & = -  \sum_{i=1}^m x_j^{(i)}  \left[y^{(i)} - \sum_{k \neq j}^n \theta_k x_k^{(i)} - \theta_j x_j^{(i)}\right]
     \\
     & = -  \sum_{i=1}^m x_j^{(i)} \left[y^{(i)} - \sum_{k \neq j}^n \theta_k x_k^{(i)} \right] +  \theta_j \sum_{i=1}^m (x_j^{(i)})^2
\\
& \triangleq - \rho_j + \theta_j z_j
\end{aligned}


\begin{equation}
    \partial_{\theta_j} \lambda \sum_{j=0}^n |\theta_j|  =      \partial_{\theta_j} \lambda |\theta_j|=
    \begin{cases}
      \{ - \lambda \} & \text{if}\ \theta_j < 0 \\
      [ - \lambda , \lambda ] & \text{if}\ \theta_j = 0 \\
      \{  \lambda \} & \text{if}\ \theta_j > 0 
    \end{cases}
  \end{equation}

```python
def Lasso_regression_CD(orig_dataset, features, target, 
                        lambda_ = 1e-4, threshold = 1e-8, n_iter = 1e5, intercept = True):
    
    dataset = orig_dataset.copy()
    dataset['coef'] = 1
    
    features = (['coef'] + list(features))
    
    #print(target)
    H = dataset[features].values
    y = dataset[target].values

    #H = H / (np.linalg.norm(H,axis = 0))
    #print(H)
    try:
        H.shape[1]
    except:
        H = H.reshape((len(H), 1))
    y = y.reshape((len(y), 1))

    w = np.zeros(len(features))

    cost_prev = np.inf
    for i in range(int(5)):
        
        cost = 0
        for j in range(len(features)):
            
            #print(w)
            x_j = H[:,j].reshape(-1, 1)
            
            y_pred = H.dot(w)
            y_pred = y_pred.reshape((len(y_pred), 1))
            
            rho = x_j.T.dot(y - y_pred + w[j]*x_j)
            z = np.sum(np.square(x_j))
             
            if rho < - lambda_:
                w[j] =  (rho + lambda_)/z
            elif rho >  lambda_:
                w[j] =  (rho - lambda_)/z
            else: 
                w[j] =  0
            
            if intercept == True:  
                if j == 0: 
                    w[j] =  rho 
            
            #print(w)
            
        cost = (1/len(y)) * np.sum(np.square(H.dot(w) - y)) + sum(map(abs, w))
        
        if cost_prev - cost < threshold:
            break
        cost_prev = cost
        
    y_pred = H.dot(w)
    
    dataset['y_pred'] = y_pred
    
    print('------------------------------------------------')
    print('R2 Score: ', r2_score(dataset[target], dataset['y_pred']))
    print('MSE: ', mean_squared_error(dataset[target], dataset['y_pred']))
    print('RMSE: ', np.sqrt(mean_squared_error(dataset[target], dataset['y_pred'])))
    print('MAE: ', mean_absolute_error(dataset[target], dataset['y_pred']))
    print('------------------------------------------------')
    
    print('Features: ', features)
    print('Coefficient:', w)
    return dataset.drop('coef', axis=1), w
```

### Low Dimension Data

```python
%%time
pred_dataset, w = Lasso_regression_CD(dataset, 'x', 'y')   
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x = 'x', y='y', data=pred_dataset, ax=ax)
sns.lineplot(x = 'x', y='y_pred', data=pred_dataset, ax=ax, color='red')
plt.show()
```
